```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

norm = np.random.normal(50,100,100)
norm.sort()
plt.hist(norm, bins=20, color='b')
plt.axvline(norm.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(norm.mean() + norm.std(), color='black', linestyle='solid', linewidth=2)
plt.axvline(norm.mean() - norm.std(), color='black', linestyle='solid', linewidth=2)

```

```python
pois = np.random.poisson(10,100)
plt.hist(pois, bins=20, color='r')
plt.axvline(pois.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pois.mean() + pois.std(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pois.mean() - pois.std(), color='b', linestyle='solid', linewidth=2)
plt.show()
```

```python
logi = np.random.logistic(0,50,100)
plt.hist(logi, bins=20, color='purple')
plt.axvline(logi.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logi.mean() + pois.std(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logi.mean() - pois.std(), color='b', linestyle='solid', linewidth=2)
```

```python
power = np.random.power(5, 100)
plt.hist(power, bins=20, color='purple')
plt.axvline(power.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(power.mean() + power.std(), color='b', linestyle='solid', linewidth=2)
plt.axvline(power.mean() - power.std(), color='b', linestyle='solid', linewidth=2)
```

```python
bino = np.random.binomial(100, 0.5, 1000)
plt.hist(bino, bins=20, color = 'green')
plt.axvline(bino.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(bino.mean() + bino.std(), color='b', linestyle='solid', linewidth=2)
plt.axvline(bino.mean() - bino.std(), color='b', linestyle='solid', linewidth=2)
```

```python
gamma = np.random.gamma(5, 1, 100)
plt.hist(gamma, bins=20, color = 'green')
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='b', linestyle='solid', linewidth=2)
```

```python
var1 = np.random.normal(5,0.5,100)
var2 = np.random.normal(10,1,100)
var3 = var1 + var2
plt.hist(var3, bins=100, color='green')
plt.axvline(var3.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='black', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
```
